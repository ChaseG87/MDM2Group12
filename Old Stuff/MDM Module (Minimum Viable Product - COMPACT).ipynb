```python
## Importing Modules
import pandas as pd

'''
This is the compact version of the Minimum Viable Product - only the most core code.
Assumes correct brackets, and takes a function and applies it to another function.
It then simplifies this function.

Method:
1. Identify functions and expressions as classes
2. Evaluate application of innermost objects on each other
3. Evaluation alpha conversion, then beta, then <eta> reduction (maybe)
4. Repeat 2 and 3 until it cannot be done anymore

Areas for future expansion:
- Reading from a file
- Eta reduction
- Figuring out correct brackets
- Figuring out whether a lambda calculus statement is correct
- Diagnostics
- Multiple Lambda Calclus inputs
'''

class Function:
    def __init__(self, string):
        period_index = string.index('.')
        self.vars = string[1:period_index]
        self.body = string[period_index+1:]
        self.string = string

    def beta_reduce(self, inputs):
        '''
        inputs is a list of each individual input
        This does the beta reduction using the relative inputs.
        '''

        def alpha_conversion(input, body):
            '''
            Changes input into form that doesnt conflict with body via alpha conversion.
            '''

            def find_replacement(input, body):
                """ Finds a Substitute Variable - Used in the Alpha Reduction Function"""
                
                option_list = 'abcdefghijklmnopqrstuvwxyz'
                for option in option_list:
                    if option not in input and option not in body:
                        return option
        
            for char in input:
                if char not in '/.()' and char in body:
                    replacement = find_replacement(input, body)
                    for i in range(input.count(char)):
                        rep_index = input.index(char)
                        input = input[:rep_index] + replacement + input[rep_index+1:]
            return input


        var_list = [self.vars[i] for i in range(len(self.vars))]
        ins = ['(' + input.string + ')' if type(input) == Function else input for input in inputs]

        for index in range(len(ins)):           
            var = var_list[index]
            self.vars = self.vars[1:]
            innie = ins[index]
            innie = alpha_conversion(innie, self.body)

            for i in range(self.body.count(var)):
                insert_index = self.body.index(var)
                self.body = self.body[0:insert_index] + innie + self.body[insert_index+1:]

## RUNNING CODE
sent1 = Function('/xy.xxyy')
sent2 = 'yy'
sent3 = Function('/bc.c')

sent1.beta_reduce([sent3])
print(sent1.body, sent1.vars)

sent1.beta_reduce([sent2])
print(sent1.body, sent1.vars)
```
